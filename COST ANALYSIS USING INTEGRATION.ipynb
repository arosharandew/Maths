```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
```

```python
t, x = sp.symbols('t x')
```

```python
F_bus_int = 2*t + 1
F_taxi_int = 3*t + 2
```

```python
F_bus = sp.integrate(F_bus_int, (t, 0, x))
F_taxi = sp.integrate(F_taxi_int, (t, 0, x))

print("Bus fare function:")
print("F_b(x) =", F_bus)
print()
print("Taxi fare function:")
print("F_t(x) =", F_taxi)
print()
```

```python
print("Integration working:")
print("Bus: integral from 0 to x of (2t + 1) dt")
print("= [t^2 + t] from 0 to x")
print("= (x^2 + x) - (0 + 0)")
print("= x^2 + x")
print()
print("Taxi: integral from 0 to x of (3t + 2) dt")
print("= [(3/2)t^2 + 2t] from 0 to x")
print("= (1.5*x^2 + 2*x) - (0 + 0)")
print("= 1.5*x^2 + 2*x")
print()
```

```python
x_vals = np.linspace(0, 5, 100)

bus_fare = x_vals**2 + x_vals
taxi_fare = 1.5*x_vals**2 + 2*x_vals
```

```python
x_vals = np.linspace(0, 5, 100)

bus_fare = x_vals**2 + x_vals
taxi_fare = 1.5*x_vals**2 + 2*x_vals
```

```python
plt.figure(figsize=(10, 6))

plt.plot(x_vals, bus_fare, 'b-', linewidth=2, label='Bus Fare')
plt.plot(x_vals, taxi_fare, 'r-', linewidth=2, label='Taxi Fare')

plt.xlabel('Distance (km)')
plt.ylabel('Fare (Rupees)')
plt.title('Bus vs Taxi Fares in Colombo')
plt.grid(True)
plt.legend()

plt.show()
```

```python
print("Fare comparison for Fort to Borella (4.5 km):")
bus_fare_45 = 4.5**2 + 4.5
taxi_fare_45 = 1.5*4.5**2 + 2*4.5

print("Bus fare: Rs", round(bus_fare_45, 2))
print("Taxi fare: Rs", round(taxi_fare_45, 2))
print("Difference: Rs", round(taxi_fare_45 - bus_fare_45, 2))
print()
```

```python
print("Short trip (2 km):")
bus_short = 2**2 + 2
taxi_short = 1.5*2**2 + 2*2
print("Bus: Rs", bus_short, "Taxi: Rs", taxi_short)
print("Taxi is", round(taxi_short/bus_short, 1), "times more expensive")
print()

print("Long trip (5 km):")
bus_long = 5**2 + 5
taxi_long = 1.5*5**2 + 2*5
print("Bus: Rs", bus_long, "Taxi: Rs", taxi_long)
print("Taxi is", round(taxi_long/bus_long, 1), "times more expensive")
print()
```

```python
print("Affordability analysis:\n"
      "Short trips: Taxi is relatively more expensive\n"
      "Long trips: Taxi becomes even more expensive\n"
      "Bus is always more affordable for Colombo travel\n"
      "Integration shows fare growth patterns clearly")
```
